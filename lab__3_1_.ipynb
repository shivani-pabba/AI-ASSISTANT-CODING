###**1: Zero-Shot Prompting (Palindrome Number Program)**

### #Write a Python function that checks whether a given number is a palindrome.

**AI-Generated Code**

```python
def is_palindrome(num):
    num = str(num)
    if num == num[::-1]:
        return True
    return False
```

##### **Observation:** The program correctly checks palindrome numbers by reversing the input and comparing it with the original value.


###**2: One-Shot Prompting (Factorial Calculation)**

### #Write a Python function to compute the factorial of a given number.
### Example: Input: 3 -> Output: 6

**AI-Generated Code:**

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact
```

**Testing Outputs:**

Input: 5  
Output: 120

Input: 1  
Output: 1

Input: -3  
Output: Invalid input

**Observation:** The program finds the factorial by multiplying numbers from 1 to the given number and shows an error for negative input.

###**3:Few-Shot Prompting (Armstrong Number Check)**

## #Write a Python function to check whether a given number is an Armstrong number.
###Examples:

Input: 153 → Armstrong Number

Input: 370 → Armstrong Number

Input: 123 → Not an Armstrong Number

**AI-Generated Code:**

```python
def is_armstrong(num):
    if num < 0:
        return "Invalid input"
    digits = len(str(num))
    total = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10
    if total == num:
        return "Armstrong Number"
    else:
        return "Not an Armstrong Number"
```

**Testing Outputs:**

Input: 153  
Output: Armstrong Number

Input: 370  
Output: Armstrong Number

Input: 123  
Output: Not an Armstrong Number

**Observation:** An Armstrong number is a number whose value is equal to the sum of the cubes of its digits.

###**4:Context-Managed Prompting (Optimized Number Classification)**

### #You are an experienced Python developer.Write an optimized Python program to classify a number as prime, composite, or neither.Ensure input validation and efficient logic.

**AI-Generated Code:**

```python
def classify_number(n):
    if n <= 1:
        return "Neither Prime nor Composite"
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return "Composite"
    return "Prime"
```

Output: Composite

**Observation:** The program classifies a number as prime, composite, or neither by checking its divisibility.

##**5: Zero-Shot Prompting (Perfect Number Check)**


##  #Write a Python function that checks whether a given number is a perfect number.

**AI-Generated Code:**

```python
def is_perfect(num):
    if num <= 0:
        return False
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    return total == num
```

Output: True

Output: True

Input: 12

Input: 0

Input: -5

**Observation:** The program checks whether a number is perfect by adding all its factors and comparing the sum with the original number.

###**6: Few-Shot Prompting (Even or Odd Classification with Validation)**

## #Write a Python program to determine whether a number is even or odd with proper input validation.

###Examples:

Input: 8 → Even  
Input: 15 → Odd  
Input: 0 → Even

```python
def even_or_odd(num):
    if not isinstance(num, int):
        return "Invalid input"
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

**Testing Outputs:**

Input: 15 → Odd  
Input: 0 → Even  
Input: -4 → Even  
Input: 3.5 → Invalid input


**Observation:** The program checks whether a number is even or odd and shows invalid input for non-integer values.
